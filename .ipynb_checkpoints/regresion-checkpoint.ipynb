```python
MODELOS DE APRENDIZAJE AUTOMATICO
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import MinMaxScaler
from sklearn.naive_bayes import GaussianNB
```

```python
#Dividir datos en tren y datos de prueba
X = data.drop('diagnosis', axis= 1)
y = data['diagnosis']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state= 42)
```

```python
#Normalizacion de datos
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Modelo Naive Bayes
naive = GaussianNB()
#entrenando el modelo
naive.fit(X_train, y_train)
```

```python
#PredicciÃ³n de valores con el modelo
y_nb_predict = naive.predict(X_test)
```

```python
#Metricas de modelo de los datos de prueba
f'Accuracy: {round(accuracy_score(y_test, y_nb_predict) * 100, 2)}%'
```

```python
#Matrix ce confusion
ax = sns.heatmap(confusion_matrix(y_test, y_nb_predict), annot= True, fmt= 'g')
ax.set_title('Matriz de confusiÃ³n de datos predichos a partir de datos de prueba')
plt.show()
```

```python
#Informe de clasificacion
print(classification_report(y_test, y_nb_predict))
```

```python
#De los datos del tren
y_nb_train_predict = naive.predict(X_train)
```

```python
f'Accuracy: {round(accuracy_score(y_train, y_nb_train_predict) * 100, 2)}%'
```

```python
#Matrix ce confusion
ax = sns.heatmap(confusion_matrix(y_train, y_nb_train_predict), annot= True, fmt= 'g')
ax.set_title('Matriz de confusiÃ³n de datos pronosticados a partir de datos de entrenamiento')
plt.show()
```

```python
#Informe de clasificacion
print(classification_report(y_train, y_nb_train_predict))
```

```python
#ValidaciÃ³n cruzada
from sklearn.model_selection import KFold, cross_val_score
```

```python
kfold = KFold(n_splits= 30, shuffle= True, random_state= 42)
model = GaussianNB()
result = cross_val_score(model, X_train, y_train, cv= kfold)
```

```python
f'Cross validation for Naive Bayes model: {round(result.mean() * 100, 2)}%'
```

```python
MODELO DE REGRESION LINEAL


```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(random_state= 42, max_iter= 100, tol= 0.0001, penalty= 'l2', C= 1, solver= 'lbfgs')
#Entrenando modelo de regresion logistica
lr.fit(X_train, y_train)
```

```python
y_lr_predict = lr.predict(X_test)
```

```python
METRICAS DE REGRESION LOGISTICA
```

```python
f'Accuracy: {round(accuracy_score(y_test, y_lr_predict) * 100, 2)}%'
```

```python
#Matrix ce confusion
ax = sns.heatmap(confusion_matrix(y_test, y_lr_predict), annot= True, fmt= 'g')
ax.set_title('Matriz de confusiÃ³n de datos pronosticados a partir de datos de prueba')
plt.show()
```

```python
#Informe de clasificacion
print(classification_report(y_test, y_lr_predict))
```

```python
#ValidaciÃ³n cruzada
model = LogisticRegression(random_state= 42, max_iter= 100, tol= 0.0001, penalty= 'l2', C= 1, solver= 'lbfgs')
result = cross_val_score(model, X_train, y_train, cv= kfold)
```

```python
f'Cross validation for Logistic Regression model: {round(result.mean() * 100, 2)}%'
```
