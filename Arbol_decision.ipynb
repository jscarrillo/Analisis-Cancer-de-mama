## Se Cargan las librerias y se cargan datos

```python
import pandas as pd
import matplotlib.pyplot as plt
data= pd.read_excel ('cancer_mama.xlsx')
data.head()
```

```python
data.info()
```

```python
x = data.iloc [:,2:33]
y = data.iloc[:,1]
x.head()
```

```python
from sklearn.model_selection import train_test_split
#x_train y y Train para entrenamiento
#y_train y y_test para prueba
X_train, X_test, Y_train, Y_test = train_test_split(x, y, train_size = 0.75, random_state = 0)
```

```python
X_train.info()
```

## Se crea el modelo del arbol de decision

```python
from sklearn.tree import DecisionTreeClassifier

arbol =  DecisionTreeClassifier ()
```

## Entrenamos el modelo

```python
Arbol_Diagnostico = arbol.fit (X_train, Y_train)
```

## Graficamos el arbol de decision

```python
from sklearn import tree

fig = plt.figure (figsize = (20, 25))

tree.plot_tree(Arbol_Diagnostico, feature_names = list (x.columns.values), class_names = list (y.values), filled = True)

plt.show()
```

## Se hace la prediccion con los datos contenidos en la tabla de testing

```python
y_pred = Arbol_Diagnostico.predict (X_test)

y_pred
```

```python
from sklearn.metrics import confusion_matrix

matriz_Confusion =  confusion_matrix (Y_test, y_pred)

matriz_Confusion
```

## *La cantidad de los valores reales para la clase benigno es 90 y la cantidad de los valores reales para la clase maligno es 53 - Filas

## *La cantidad de los valores predichos para la clase benigno es 84 y la cantidad de los valores predichos para la clase maligno es 59 - Columnas

## *La cantidad de los valores correctamente predichos es 129 y la cantidad de los valores incorrectamente predichos es 14 - Diagonales

## A CONTINUACION CALCULAMOS LA PRESICION GLOBAL DEL MODELO

```python
import numpy as np

Precision_Global = np.sum (matriz_Confusion.diagonal())/np.sum (matriz_Confusion) 
Precision_Global
```

## TAMBIEN CALCULAMOS LA PRECISION POR CADA CLASE 

```python
Precision_B = ((matriz_Confusion [0,0]))/sum (matriz_Confusion[0,])
Precision_B
```

```python
Precision_M = ((matriz_Confusion [1,1]))/sum (matriz_Confusion[1,])
Precision_M
```
